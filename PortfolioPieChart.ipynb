```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.express as px

```

```python


x = [15.8 , 9.6 , 8.5 , 8.1 , 6.9 , 6.6, 6.1, 3.3]
MyLabels = ['PETR4', 'VALE3', 'BBSE3', 'ENBR3' , 'CPLE6' , 'TAEE11', 'BBAS3','SAPR11']
MyColors = ['red', 'orange', 'green', 'gray', 'yellow','black','cyan','blue']

plt.pie(x, colors = MyColors, autopct='%0.01f%%', pctdistance=1.20)
plt.legend(labels=MyLabels, title='Stocks', bbox_to_anchor=(1.5,0,0,1,), loc='center right', prop = {'size':10})
plt.title('Portfolio', size='18', color='black')
plt.show() 


```

```python
dados_x = ['PETR4', 'VALE3','BBSE3','ENBR3','CPLE6','BBAS3']
dados_y = [12.9 , 9.1 ,7.8, 7.6 , 6.0 , 5.3 ]

fig = px.pie(names=dados_x , values=dados_y, title='Portfolio')
fig.update_traces(textposition='inside', textinfo='percent+label')

fig.show()

```
